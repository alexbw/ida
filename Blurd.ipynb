```python
import skimage
import Image
from scipy import ndimage
```

```python
# Load the image
img = np.array(Image.open("imgs.png"))
```

```python
# Identify the high-points of the signal
def get_norep(labels):
    labels_norep = []
    durations = []
    labels_norep.append(labels[0])
    durations.append(1)
    for i in range(1,len(labels)):
        if labels[i] == labels_norep[-1]:
            durations[-1] += 1
        else:
            labels_norep.append(labels[i])
            durations.append(1)
    return np.array(labels_norep), np.array(durations)


def find_highs(data, thresh=100, min_length=15):
    highs, durations = get_norep((data>thresh).astype('int32'))
    idx = np.argwhere((durations>min_length)&(highs==1)).ravel()
    cdur = np.r_[0,np.cumsum(durations)]
    regions = np.vstack([cdur[idx],cdur[idx+1]]).T
    return regions
```

```python
# Downsample it
r = []
for i in range(3):
    r.append(ndimage.zoom(img[:,:,i], 0.05))
r = np.dstack(r)
rs = r.mean(2)
```

```python
figure()
thresh = 100
h = ndimage.median_filter(np.max(rs,axis=1), (rs.shape[1]/10.0,))
dh = np.abs(np.diff(h))
w = ndimage.median_filter(np.max(rs,axis=0), (rs.shape[0]/50.0,))
dw = np.abs(np.diff(w))
plot(h)
plot(dh-200)
figure()
plot(h>thresh)
figure()
plot(w)
plot(dw-200)
```

```python
figure(figsize=(3,15))
imshow(rs, cmap='Greys_r')
```

```python
height_ticks = find_highs(h,min_length=100)
width_ticks = find_highs(w,min_length=10)
```

```python
figure(figsize=(3,15))
imshow(rs, cmap='Greys_r')
for height_tick in height_ticks:
    for width_tick in width_ticks:
        vlines(width_tick[0], height_tick[0], height_tick[1], linewidth=2, color='r')
        vlines(width_tick[1], height_tick[0], height_tick[1], linewidth=2, color='r')
        hlines(height_tick[0], width_tick[0], width_tick[1], linewidth=2, color='r')
        hlines(height_tick[1], width_tick[0], width_tick[1], linewidth=2, color='r')
```

```python
# Extract the sub-images
test_imgs_bw = []
test_imgs = []
for height_tick in height_ticks:
    h1,h2 = height_tick
    for width_tick in width_ticks:
        w1,w2 = width_tick
        Ibw = rs[h1:h2,w1:w2]
        test_imgs_bw.append(Ibw)
        I = r[h1:h2,w1:w2,:]
        test_imgs.append(I)
```
