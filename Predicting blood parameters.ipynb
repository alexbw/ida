```python
def classify(data, y, n_folds=5, C=50, gamma=1.0):
    from sklearn.linear_model import LogisticRegression
    from sklearn.svm import SVC
    learner = SVC(C=C, gamma=gamma, shrinking=True)
    precisions = []
    recalls = []
    y_scores = []
    all_test_idx = []
    all_pred_y = []
    for train_idx, test_idx in cross_validation.Bootstrap(len(data), n_iter=n_folds, train_size=0.5):
        all_test_idx.append(test_idx)
        learner.fit(data[train_idx], y[train_idx])
        pred_y = learner.predict(data[test_idx])
        all_pred_y.append(pred_y)
        y_scores += [learner.decision_function(data[test_idx])]
        precision, recall, fscore, support = metrics.precision_recall_fscore_support(y[test_idx], pred_y, average='weighted')
        precisions.append(precision)
        recalls.append(recall)
        
    precisions = np.array(precisions)
    recalls = np.array(recalls)
    
    accuracy = cross_validation.cross_val_score(learner, data, y, cv=10, )
    print "Precision: %f, Recall: %f" % (precisions.mean(), recalls.mean())
    all_true_y = [y[test_idx] for test_idx in all_test_idx]
    return all_pred_y, all_true_y, all_test_idx, y_scores
```

```python
def regress(data, y, n_folds=50,gamma=1.0, epsilon=0.1):
    from sklearn.linear_model import LinearRegression, Lasso
    from sklearn.svm import SVR
#     learner = LinearRegression(fit_intercept=True)
    learner = SVR('rbf', gamma=gamma, epsilon=epsilon)
    precisions = []
    recalls = []
    y_scores = []
    all_test_idx = []
    all_pred_y = []

    for train_idx, test_idx in cross_validation.KFold(len(data), shuffle=False, n_folds=n_folds):
        all_test_idx.append(test_idx)
        learner.fit(data[train_idx], y[train_idx])
        pred_y = learner.predict(data[test_idx]).ravel()
        all_pred_y.append(pred_y)
        y_scores += [learner.score(data[test_idx], y[test_idx])]
        
    accuracy = cross_validation.cross_val_score(learner, data, y, cv=10)
    all_true_y = [y[test_idx].ravel() for test_idx in all_test_idx]
    print "Avg R2: %2.2f" % (np.mean(y_scores))
    diffs = np.hstack(all_pred_y).ravel() - np.hstack(all_true_y).ravel()
    print "Std Diff: %2.2f" % (np.std(diffs))
    return all_pred_y, all_true_y, all_test_idx, y_scores
```
