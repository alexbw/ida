```python
import pandas as pd
import numpy as np
```

```python
sdf = pd.load("data.pandas")
```

```python
# Try many different combinations, to check all data
isdf = sdf[(sdf.blood_type == 'tt') & (sdf.ida == "IDA3") & (sdf.run_time == 4) & (sdf.good == True)]
```

```python
# View the data properties, to go back to the original file
for i in range(len(isdf)):
    print str(i) + ":"
    
    print "\t" +  "/".join(isdf.tiff_file.values[i].split("/")[-2:])
    print "\t" + str(isdf.tiff_file_position.values[i])
    print "\t" + str(isdf.good.values[i])
    print "\t" + str(isdf.sample.values[i])
```
