## We first down-sample the image by a factor of 3 (the scans are large)

```python
# Downsample it
channels = []
for i in range(3):
    channels.append(ndimage.zoom(img[:,:,i], 0.05))
channels = np.dstack(channels)
bw_img = channels.mean(2)
```

```python
figure(figsize=(20,4))
imshow(bw_img.T, cmap='Greys_r')
axis('tight')
axis('off')
```

## We find the light/dark boundaries of the tubes in height and width.

These boundaries mark the edges of the tube images

```python
# Extract light/dark boundaries
h = ndimage.median_filter(np.max(bw_img,axis=1), (bw_img.shape[1]/10.0,))
w = ndimage.median_filter(np.max(bw_img,axis=0), (bw_img.shape[0]/50.0,))
```

```python
figure(figsize=(20,4))
gs = GridSpec(2,2, width_ratios=[10,1], height_ratios=[3,10], wspace=0.0, hspace=0.1)

# Show the image
subplot(gs[1,0])
imshow(bw_img.T, cmap='Greys_r')
axis('tight')
axis('off')

# Show the height (transposed, so it lies horizontally) markers
thresh = 100
dh = np.abs(np.diff(h))
dw = np.abs(np.diff(w))

subplot(gs[0,0])
plot(h)
plot(dh-200)
ylim(-300,350)
# axis('tight')
xlim(0,len(h))
axis('off')

subplot(gs[1,1])
# Show the width markers (again, transposed, so it's vertical)
# figure()
# plot(h>thresh)
# figure()
plot(w,range(len(w)))
plot(dw-200,range(len(dw)))
axis('tight')
axis('off')
xlim(-300, 350)

# plot(dw-200)
```

## We extract the tubes from within these edges.

```python
height_ticks = find_highs(h,min_length=100)
width_ticks = find_highs(w,min_length=10)
```

```python
figure(figsize=(20,4))
imshow(bw_img.T, cmap='Greys_r')
grid('off')
axis('off')
for height_tick in height_ticks:
    for width_tick in width_ticks:
        hlines(width_tick[0], height_tick[0], height_tick[1], linewidth=2, color='r')
        hlines(width_tick[1], height_tick[0], height_tick[1], linewidth=2, color='r')
        vlines(height_tick[0], width_tick[0], width_tick[1], linewidth=2, color='r')
        vlines(height_tick[1], width_tick[0], width_tick[1], linewidth=2, color='r')
```

```python
# Extract the sub-images
test_imgs_bw = []
test_imgs = []
for height_tick in height_ticks:
    h1,h2 = height_tick
    for width_tick in width_ticks:
        w1,w2 = width_tick
        Ibw = bw_img[h1:h2,w1:w2]
        test_imgs_bw.append(Ibw)
        I = channels[h1:h2,w1:w2,:]
        test_imgs.append(I)
```

```python
figure(figsize=(10,5))
gs = GridSpec(1,len(test_imgs))
for i,test_img in enumerate(test_imgs):
    subplot(gs[i])
    imshow(test_img); axis('off')
```

## We extract the "redness" of the tubes

```python
saturation_threshold = 0.5
figure(figsize=(10,5))
gs = GridSpec(1,len(test_imgs))
redness_imgs = []
for i,test_img in enumerate(test_imgs):
    subplot(gs[i])
    
    # Convert the tube image to HSV colorspace, select out the saturation and values channels
    img_lab = skimage.color.rgb2hsv(test_img).astype('float32')
    saturation_img = img_lab[:,:,1]
    value_img = img_lab[:,:,2]
    
    # Set all pixels in the V (value) channel to 1 if their corresponding 
    # pixel in the S (saturation) channel is valued less than 0.5
    value_img[saturation_img < saturation_threshold] = 1.0
    SoverV = saturation_img/value_img
    
    # Return redness as the S channel / V channel (the only saturated 
    # color in the tube images is red, so we are in effect extracting the intensity of any color)
    imshow(SoverV); axis('off')
    redness_imgs.append(SoverV)
```

## The plug in the tube contains no relevant information, so we automatically identify it and crop it out

Also, we sorted the dataset by total luminance in the red channel, and manually identified tubes that contained no blood, and removed them from the dataset.

```python
figure(figsize=(10,5))
gs = GridSpec(1,len(test_imgs))
# Threshold for the rate of drop-off required to identify the plug
threshold = 0.1
# Don't look for the plug in all locations of the tube, just in the top 15% or so.
valid_range=(3,25)

cropped_imgs = []
data_traces = []
for i,red_img in enumerate(redness_imgs):
    subplot(gs[i])
    
    # To find the plug, first calculate the mean across the width of the tube
    data_trace = red_img.mean(1)
    
    # Then find the position of the plug as the location of fastest drop-off
    # in the "redness" trace.
    d = np.diff(data_trace)
    d[d<threshold] = 0.0
    peaks = np.argwhere(local_maxima(d)).ravel()
    peaks = peaks[(peaks > valid_range[0]) & (peaks < valid_range[1])]
    plug_position = peaks[np.argmax(d[peaks]).ravel()[0]].ravel()[0]-1
    imshow(red_img); axis('off')
    hlines(plug_position, 0, red_img.shape[1], color='r')
    
    cropped_imgs.append(red_img[plug_position:])
    data_traces.append(data_trace[plug_position:])
```

## Finally, 1 dimensional data associated with each tube, and save it for analysis.

```python
figure(figsize=(10,5))
gs = GridSpec(1,2*len(test_imgs))
for i,(data_trace, cropped_img) in enumerate(zip(data_traces, cropped_imgs)):
    subplot(gs[i*2])
    imshow(cropped_img); axis('tight'); axis('off')
    subplot(gs[(i*2)+1])
    plot(data_trace[::-1], range(len(data_trace)))
    axis('tight'); axis('off')
```
